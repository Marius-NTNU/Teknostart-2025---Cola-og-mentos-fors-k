```python
#vi må starte med å importere noen pakker som ikke er med i standardbiblioteket til python
import matplotlib.pyplot as plt
import numpy as np
import scipy.optimize
```

Vi skal lage en modell som beskriver banen til colaen når den skytes ut av flasken. Man kan bruke en andregradslikning og randbetingelsene som er beskrevet i figuren nedenfor.  



i tillegg skal vi bruke disse formlene: 

sin$\theta$ = $\frac{\text{motstående katet}}{\text{hypotenus}}$, 
cos$\theta$ = $\frac{\text{hosliggende katet}}{\text{hypotenus}}$

$y(x) = ax^2+bx+c$

Vi skal bruke en numerisk metode for å regne ut koeffisientene a, b, og c til polynomet.

Vi starter med å definere noen konstanter. Dette er verdier vi har målt og som ikke modellen trenger å regne ut. 

```python
L = 5 #[m] skriv inn lengden på colastrålen dere målte
theta = 40 #[deg] dette er vinkelen dere brukte på colaflasken
theta_rad = theta*(np.pi/180) #[rad]
flaske = 0.317 #[m] dette er høyden på colaflasken
```

Ettersom vi har tre ukjente må vi finne tre likninger for å kunne løse likninssettet. 
Vi definerer tre punkt som vi har kjenskap til på kurven og bruker disse for å finne de ukjente koeffisientene. 
Det er viktig at disse likningene skrives på formen f = 0 ettersom vi skal bruke en funksjon senere som tar inn likninger på denne formen.

```python
def model(guesses):
    a, b, c = guesses
    f0 = - y0 + a*x0**2 + b*x0 + c 
    f1 = - y1 + a*x1**2 + b*x1 + c 
    fL = - yL + a*xL**2 + b*xL + c 
    return [f0, f1, fL]
```

Vi har verdier for y0, x0, yL, og xL. Vi må altså finne utrykk for y1 og x1. Vi kan utnytte at andregradslikningen er tilnærmet lik lineær når man er nær begynnelsen av kurven.
Dersom vi antar at at brusen skytes ut rett fra halsen til flasken kan man finne høyden til flaskehalsen ved å bruke sinus til vinkelen til flasken (se illustrasjon), dette er y1. x1 er cosinus til vinkelen.

```python
y1 = np.sin(theta_rad)*flaske
x1 = np.cos(theta_rad)*flaske

x0 = 0
y0 = 0

xL = L
yL = 0
```

Verdiene som ble regnet ut over kan brukes for å finne a, b, og c i likningene.
Først må vi gjette på hva disse verdiene kan være.

```python
a_guess = -1
b_guess = 1
c_guess = 0.01
guesses = [a_guess, b_guess, c_guess ]

solution = scipy.optimize.root(model, guesses) # Denne funksjonen bruker numerikk for å finne roten til utrykkene over ved å ta utgangspunkt i gjettene vi har oppgitt.
a = solution.x[0]
b = solution.x[1]
c = solution.x[2]
print('a:', a)
print('b:', b)
print('c:', c)
```

Ved å lage en modell som beskriver banen til brusen kan vi hente ut informasjon som ikke er mulig/vanskelig å måle. Hvis vi for eksempel lurer på hva som er den maksimale høyden til strålen kan dette gjøres ved å finne toppunktet på grafen. 

```python
#f'(x) = -a*2*x+b
x_05L= xL/2
x_halv_L = b/(-a*2) #x = L/2
y_halv_L = a*x_05L**2 + b*x_halv_L + c

print(y_halv_L)
```

Vi kan også visualisere banen ved å plotte modellen i en graf. Dette gir en visuell representasjon av banen til brusen.

```python
x = np.linspace(0,L)
y = a*x**2 + b*x + c

fig, ax = plt.subplots(figsize=(6, 3))
ax.set_xlabel('Lengde [m]')
ax.set_ylabel('Høyde [m]')
ax.plot(x,y)
plt.show()
```
